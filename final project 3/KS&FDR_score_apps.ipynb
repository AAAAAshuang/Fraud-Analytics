```python
variables=mydf.drop(['fraud_label','record','date','ssn','firstname','lastname',\
                            'address','zip_5','dob','homephone','dob_str','zip5_str',\
                               'nameDOB','fulladdress',],axis=1).columns
```

```python
mydf = mydf.drop(['record','date','ssn','firstname','lastname',\
                'address','zip_5','dob','homephone','dob_str','zip5_str',\
                    'nameDOB','fulladdress',],axis=1)
```

```python
mydf.shape
```

```python
nvars=len(variables)
print(nvars)
```

```python
list(mydf.columns)
```

```python
goods=mydf[mydf['fraud_label']==0]
goods.shape
```

```python
bads=mydf[mydf['fraud_label']==1]
bads.shape
```

```python
numgoods=len(goods)
numbads=len(bads)
```

```python

KS=pd.DataFrame(np.zeros((nvars+2,3)))
i=0
for column in mydf:
    print(i,column)
    KS.iloc[i,0]=column
    i=i+1
KS
```

```python
KS.columns = ['field','ks','FDR']
KS
```

```python
n=0
for i in mydf.columns:
    x = goods[i]
    y = bads[i] 
    KS['ks'][n] = sps.ks_2samp(x, y)[0]
    n+=1
```

```python
KS.sort_values(by=['ks'],ascending=False,inplace=True)
KS
```

```python
topRows= int(round(len(mydf)*0.03))
j=0
for column in mydf:
    temp1=mydf.sort_values(column,ascending=False).head(topRows)
    temp2=mydf.sort_values(column,ascending=True).head(topRows)
    needed1=temp1.loc[:,'fraud_label']
    needed2=temp2.loc[:,'fraud_label']
    FDR1=sum(needed1)/numbads
    FDR2=sum(needed2)/numbads
    FDRate=np.maximum(FDR1,FDR2)
    #print(j,column,FDR1,FDR2,FDRate)
    KS.loc[j,'FDR']=FDRate
    j=j+1
```

```python
KS.sort_values(by=['ks'],ascending=False,inplace=True)
KS.to_csv('KS.csv')
KS.sort_values(by=['FDR'],ascending=False,inplace=True)
KS.to_csv('FDR.csv')
KS['rank_ks']=KS['ks'].rank(ascending=True)
KS['rank_FDR']=KS['FDR'].rank(ascending=True)
KS.head(20)
```

```python
KS['average_rank']=(KS['rank_ks']+KS['rank_FDR'])/2
KS.sort_values(by=['average_rank'],ascending=False, inplace=True)
KS.reset_index(drop=True)
```
