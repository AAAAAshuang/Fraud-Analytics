```python
applications['dob_str']= applications['dob'].astype(str)
applications['zip5_str'] = applications['zip5'].astype(str)
```

```python
applications['nameDOB'] = applications[['firstname', 'lastname', 'dob_str']].apply(lambda x: '_'.join(x), axis=1)
applications['fulladdress'] = applications[['address', 'zip5_str']].apply(lambda x: '_'.join(x), axis=1)
```
